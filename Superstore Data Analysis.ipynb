4. Visualization Of Data

Frequency Of Categories

```python
fig, ax = plt.subplots(figsize=(6, 5))
category_frequency = sns.countplot(x = 'Category', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
# The frequency of Technology item is less.
# The frequency of Furniture item is medium.
# The frequency of Office Supplies is high.
```

Pie Plot Of Sub-Category

```python
plt.figure(figsize = (12,10))
df['Sub-Category'].value_counts().plot.pie(autopct = "%1.1f%%")
plt.show()
```

```python
# Binders and Paper items of Office Supplies dominate in Sales.
```

Sale vs Profit For Each Sub-Category

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title("Sales vs Profit per Sub-Category")
plt.legend(title = "Sales vs Profit")
plt.rcParams['figure.figsize'] = [12,10]
plt.show()
```

```python
# Profit is most earned through Copiers.
# But Sales is more in Chairs & papers.
# People don't buy Bookcases & Tables form the superstore, which leads to a loss in these departments.
```

Count Of Cities

```python
plt.figure(figsize = (12,10))
sns.countplot(x = df['State'])
plt.xticks(rotation = 90)
plt.show()
```

```python
# More cities in California.
# New York with moderate count of cities.
# Less cities in Wyoming & West Virginia.
```

Sale vs Profit For Each Region

```python
df.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title("Sales vs Profit per Region")
plt.legend(title = "Sales vs Profit")
plt.rcParams['figure.figsize'] = [12,10]
plt.show()
```

```python
# The Sales in the regions descend in the order : West, East, Central, South.
# The Profits in the regions descend in the order : West, East, South, Central.
# Profit in the Central is less regardless of bigger sales than the South.
```

Heat Map Of Correlation Among The Columns

```python
fig, ax = plt.subplots(1, 1, figsize = (8,6))
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
# Negative Correlations need to be delt with.
```

Heat Map Of Covariance Among The Columns

```python
fig, ax = plt.subplots(1, 1, figsize = (8,6))
sns.heatmap(df.cov(), annot = True)
plt.show()
```

```python
# Negative Covariance : (Sales, Discount), (Profit, Discount).
```

Count Of Segments

```python
plt.figure(figsize = (6,5))
sns.countplot(x = df['Segment'])
plt.show()
```

Bar Plot Of Sub-Category vs Profit

```python
plt.figure(figsize = (12,10))
sns.barplot(x = df['Sub-Category'], y = df['Profit'])
plt.xticks(rotation = 90)
plt.show()
```

```python
# Loss in products : Bookcases, Tables, Supplies & Machines.
# Huge profits in Copiers.
```

Visualization Of Negative Correlation Between Discount & Profit

```python
plt.figure(figsize = (10,3))
sns.lineplot(x = 'Discount', y = 'Profit', data = df, color = 'r', label = 'Discount')
plt.legend()
```

```python
# The plot descends showing the correlation between profit and discount to be negative.
```
